```python
x=x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11
y=y1,y2,y3,y4,y5,y6,y7,y8,y9,y10,y11
z=z1,z2,z3,z4,z5,z6,z7,z8,z9,z10,z11
Dx=[]
Dy=[]
Dz=[]
time = np.arange(11)
i=0
for dx,dy in zip(x,y):
    dx=(np.average((x[i])**2))+(np.average((x0)**2))-(2*np.average(x[i]*x0))
    dy=(np.average((y[i])**2))+(np.average((y0)**2))-(2*np.average(y[i]*y0))
    dz=(np.average((z[i])**2))+(np.average((z0)**2))-(2*np.average(z[i]*z0))
    i+=1
    Dx.append(dx)
    Dy.append(dy)
    Dz.append(dz)
plt.xlabel('Time(t)')
plt.ylabel('D(t)')
plt.plot (time,Dx,'o--',color='blue',label='x')    
plt.plot (time,Dy,'o--',color='red',label='y')    
plt.plot (time,Dz,'o--',color='black',label='z')    
plt.legend()
plt.savefig('D(t)vsTime2.png',dpi=300)
```

```python
dDxdt=np.gradient(time,Dx)
dDydt=np.gradient(time,Dy)
dDzdt=np.gradient(time,Dz)
plt.xlabel('Time(t)')
plt.ylabel('D(t)')
plt.plot (time,dDxdt,'o-',color='blue',label='x')    
plt.plot (time,dDydt,'o-',color='red',label='y')    
plt.plot (time,dDzdt,'o-',color='black',label='z')    
plt.legend()
plt.savefig('D(t)vsTime3.png',dpi=300)
```
